```python
with open("fradulent_emails.txt", "r") as file:
    lines = file.readlines()
```

```python
def split_emails(lines):
    emails = []  # List to store all emails
    current_email = []  # List to store lines of the current email

    for line in lines:
        # Check if the line indicates the start of a new email
        if line.startswith("From r") and len(current_email) > 0:
            # If we're already collecting an email, start a new one
            emails.append(current_email)
            current_email = [line]  # Start with the current line
        else:
            # Otherwise, just add the line to the current email
            current_email.append(line)
    
    # Don't forget to add the last email if the file doesn't end with a new email start
    if current_email:
        emails.append(current_email)
    
    return emails
```

```python
from datetime import datetime

def process_emails(emails):
    processed_emails = []

    for email_lines in emails:
        email_dict = {}
        body_started = False
        body_lines = []

        # Process the first line for the date
        date_line = email_lines[0]
        date_str = ' '.join(date_line.split(' ')[2:])  # Skip 'From r'
        email_dict['Date'] = datetime.strptime(date_str.strip(), '%a %b %d %H:%M:%S %Y')

        # Process the remaining lines for attributes and body
        for line in email_lines[1:]:  # Start from the second line
            if line.strip() == '':
                continue

            if not body_started:
                if ':' in line:
                    key, value = line.split(':', 1)
                    email_dict[key.strip()] = value.strip()
                    if key.strip() == 'Status':
                        body_started = True  # 'Status' marks the end of attributes
                else:
                    # If a line without ':' is encountered before 'Status', it's unexpected
                    print("Unexpected format encountered in email attributes.")
            else:
                body_lines.append(line.strip())

        email_dict['Body'] = '\n'.join(body_lines)
        processed_emails.append(email_dict)

    return processed_emails
```
